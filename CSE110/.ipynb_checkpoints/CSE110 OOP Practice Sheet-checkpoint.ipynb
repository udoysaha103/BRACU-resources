# Task-1

Design a class <strong>Patient</strong> for the given code below.<br>
•	Write a class <strong>Patient</strong>.<br>
•	Write the required constructor that takes 4 parameters and initialize the instance variables accordingly.<br>
•	Write a method <strong>print_details</strong> that prints as shown in the output.<br>
•	For BMI, the formula is BMI = weight/ height^2, where weight is in kg and height in meters.<br>

<br>
COPY THIS PORTION:<br>
--------------------------------------<br>
<em>#Write your code here<br><br>
#Do not change the following code:</em><br>

```python
#Write your code here
class Patient:
    def __init__(self, name, age, weight, height):
        self.name = name
        self.age = age
        self.weight = weight
        self.height = height
    
    def print_details(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Weight:", self.weight)
        print("Height:", self.height)
        print("BMI:", self.weight / ((self.height / 100)**2))
# TODO

# Do not change the following code:
p1 = Patient("A", 55, 63.0, 158.0)
p1.print_details()
print()
p2 = Patient("B", 53, 61.0, 149.0)
p2.print_details()
```

# Task-4

•	Create a <strong>Bank Account</strong> class with Account Number, Account Name, Pin and Total Balance as instance attribute .<br>
•	Create <strong>Withdraw </strong> method that will take the amount from the user and than reduce it from the balance.<br>
•	Create <strong>Deposit</strong> method that will take the amount from the user and than add it to the balance.<br>
• Remember to show the total balance after each deposit or withdrawal made by the user.

<br>
COPY THIS PORTION:<br>
--------------------------------------<br>
<em>#Write your code here<br><br>
#Do not change the following code:</em><br>
Tom_Holden= BankAccount("234729920000034", "Tom Holden", 45879, 7854124)<br>
Tom_Holden.deposit(34000)<br>
Tom_Holden.withdraw(2145000)<br>

--------------------------------------<br>

<strong>OUTPUT:</strong><br>

Account Number: 234729920000034<br>
Account Name: Tom Holden<br>
Pin: 45879 <br>
Total Balance: 7,854,124<br>
After deposit of 34,000 the Total Balance in the account is 7,888,124. <br>
After withdrawal of 2,145,000 the Total Balance in the account is 5,743,124. <br>

```python
# Write your code here
class BankAccount:
    def __init__(self, Account_Number, Account_Name, Pin, Total_Balance):
        self.Account_Number = Account_Number
        self.Account_Name = Account_Name
        self.Pin = Pin
        self.Total_Balance = Total_Balance

        customized_balance = list(str(self.Total_Balance))
        for index in range(len(str(self.Total_Balance)) - 3, 0, -3):
            customized_balance.insert(index, ",")
        customized_balance = "".join(customized_balance)


        print(f"Account Number: {self.Account_Number}\nAccount Name: {self.Account_Name}\nPin: {self.Pin}\nTotal Balance: {customized_balance}")

    def withdraw(self, amount):
        self.Total_Balance -= amount

        customized_balance = list(str(self.Total_Balance))
        customized_amount = list(str(amount))

        for index in range(len(str(self.Total_Balance)) - 3, 0, -3):
            customized_balance.insert(index, ",")
        for index in range(len(str(amount))-3, 0, -3):
            customized_amount.insert(index, ",")

        customized_balance = "".join(customized_balance)
        customized_amount = "".join(customized_amount)

        print(f"After withdrawal of {customized_amount} the Total Balance in the account is {customized_balance}.")

    def deposit(self, amount):
        self.Total_Balance += amount

        customized_balance = list(str(self.Total_Balance))
        customized_amount = list(str(amount))

        for index in range(len(str(self.Total_Balance)) - 3, 0, -3):
            customized_balance.insert(index, ",")
        for index in range(len(str(amount))-3, 0, -3):
            customized_amount.insert(index, ",")

        customized_balance = "".join(customized_balance)
        customized_amount = "".join(customized_amount)

        print(f"After deposit of {customized_amount} the Total Balance in the account is {customized_balance}.")


# TODO

# Do not change the following code:
Tom_Holden = BankAccount("234729920000034", "Tom Holden", 45879, 7854124)
Tom_Holden.deposit(34000)
Tom_Holden.withdraw(2145000)
```

Returned value: 6.25<br>
100 / 16 = 6.25<br>
==================<br>


```python
#Write your code here
class Calculator:
    def __init__(self, first_operand, second_operand, operator):
        self.first_operand = first_operand
        self.second_operand = second_operand
        self.operator = operator

    def calculate_button(self):
        self.result = None
        
        if self.operator == '+':
            self.result = self.first_operand + self.second_operand
        elif self.operator == '-':
            self.result = self.first_operand - self.second_operand
        elif self.operator == '*':
            self.result = self.first_operand * self.second_operand
        elif self.operator == '/':
            self.result = self.first_operand / self.second_operand
        
        return self.result
    
    def details(self):
        print(self.first_operand, self.operator, self.second_operand, "=", self.result)
# TODO

#Do not change the following code:
add = Calculator(10, 20, '+')
val = add.calculate_button()
print("Returned value:", val)
add.details()
print("==================")

sub = Calculator(val, 10, '-')
val = sub.calculate_button()
print("Returned value:", val)
sub.details()
print("==================")

div = Calculator(val, 5, '*')
val = div.calculate_button()
print("Returned value:", val)
div.details()
print("==================")

mult = Calculator(val, 16, '/')
val = mult.calculate_button()
print("Returned value:", val)
mult.details()
print("==================")
```

•	Create a <strong>Vehicle</strong> class with max_speed and mileage instance attributes .<br>
•	Create a <strong>seating_capacity</strong> method that has default value of 50.<br>
•	The default <strong>fare charge</strong> of any vehicle is seating capacity * 100 and add an extra 10% on full fare as a maintenance charge. So total fare for a vehicle instance will become the final amount = total fare + 10% of the total fare.<br>

<br>
COPY THIS PORTION:<br>
--------------------------------------<br>
<em>#Write your code here<br><br>
#Do not change the following code:</em><br>
School_bus = Vehicle("School Volvo", 180, 12)<br>
print(School_bus.seating_capacity()) <br>
print("Total Bus fare is:", School_bus.fare())<br>

--------------------------------------<br>

<strong>OUTPUT:</strong><br>

Vehicle Name: School Volvo<br>
Speed: 180 <br>
Mileage: 12<br>
The seating capacity of a bus is 50 passengers<br>
Total Bus fare is: 5500.0<br>


```python
#Write your code here
class Vehicle:
    def __init__(self, name, max_speed, mileage):
        self.name = name
        self.max_speed = max_speed
        self.mileage = mileage
        
        print(f"Vehicle Name: {self.name}\nSpeed: {self.max_speed}\nMileage: {self.mileage}")

    def seating_capacity(self, capacity = 50):
        self.capacity = capacity
        return f"The seating capacity of a bus is {self.capacity} passengers"

    def fare(self):
        self.total_fare = self.capacity * 100
        self.maintenance = self.total_fare * 10 / 100
        self.final_amount = self.total_fare + self.maintenance

        return self.final_amount
# TODO

#Do not change the following code:
School_bus = Vehicle("School Volvo", 180, 12)
print(School_bus.seating_capacity())
print("Total Bus fare is:", School_bus.fare())
```

*   CGPA = sum of individual (grade point * credit) / total credit



<em>(For example: [3.3,4] can be calculated as: <br>
CGPA = ((3.3 * 3) + (4 * 3)) / 6 <br>
Here, for each course, grade point is multiplied by 3. Total credit is the numer of courses multiplied by 3. Since the example has 2 courses, therefore a total of 6 credits)<br>

**Hints:**
Need to handle the quoatation marks in "academic standing" printing.

```python
#Write your code here
class Student:
    def __init__(self, name, s_id, department, grades):
        self.name = name
        self.id = s_id
        self.department = department
        self.grades = grades

        print(f"--------- Department: {self.department} ---------")
        print(f"Name: {self.name}, ID: {self.id}")

    def calculate_CGPA(self):
        self.cgpa = sum(self.grades) / len(self.grades)

        print(f"CGPA:", round(self.cgpa, 2))

    def print_details(self):
        if self.cgpa >= 3.8:
            self.result = "'Highest Distinction'"
            print(f"Your academic standing is {self.result}\n")
        elif self.cgpa >= 3.65:
            self.result = "'High Distinction'"
            print(f"Your academic standing is {self.result}\n")
        elif self.cgpa >= 3.5:
            self.result = "'Distinction'"
            print(f"Your academic standing is {self.result}\n")
        elif self.cgpa >= 2:
            self.result = "'Satisfactory'"
            print(f"Your academic standing is {self.result}\n")
        else:
            self.result = "cannot graduate"
            print(f"You {self.result}\n")
# TODO

#Do not change the following code:
s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
s1.print_details()

s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
s2.print_details()

s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
s3.print_details()
```
