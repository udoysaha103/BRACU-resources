```python
# array based

class stack_overflow_exception(Exception):
  '''raises stack overflow exception'''
  def __init__(self):
    super().__init__("Stack is FULL")

class stack_underflow_exception(Exception):
  '''raises stack underflow exception'''
  def __init__(self):
    super().__init__("Stack is EMPTY")


class ArrayStack:
  def __init__(self, length):
    self.size = 0
    self.stack = [None] * length
  
  def push(self, ele):
    if self.size < len(self.stack):
      self.stack[self.size] = ele
      self.size += 1
    else:
      raise stack_overflow_exception
  
  def pop(self):
    if self.size > 0:
      element_removed = self.stack[self.size - 1]
      self.stack[self.size - 1] = None
      self.size -= 1
      return element_removed
    else:
      raise stack_underflow_exception
  
  def peek(self):
    if self.isEmpty():  # empty
      raise stack_underflow_exception
    else:
      return self.stack[self.size - 1]
  
  def isEmpty(self):
    if self.size == 0:  # empty
      return True
    else:
      return False


if __name__ == "__main__":
  equation = input("Please enter the equation : ")
  stack = ArrayStack(len(equation))
  flag = True
  index = 0
  character = None
  position = None

  while index < len(equation):
    if (equation[index] == '(') or (equation[index] == '{') or (equation[index] == '['):  # for opening brackets
      try:
        stack.push(f"{equation[index]} {index}")  # pushing the character along with its index
        index += 1
      except stack_overflow_exception:
        flag = False
        break

    elif (equation[index] == ')') or (equation[index] == '}') or (equation[index] == ']'):  # for closing brackets
      try:
        data = stack.peek()
        character = data[0]
        position = data[2]

        if equation[index] == ')':
          if character != '(':
            flag = False
            break
        elif equation[index] == '}':
          if character != '{':
            flag = False
            break
        elif equation[index] == ']':
          if character != '[':
            flag = False
            break
            
        stack.pop()
        index += 1
      except stack_underflow_exception:
        flag = False
        break

    else:  # for characters apart from brackets
        index += 1
  
  # printing part --->
  if flag == False:
    if (equation[index] == '(') or (equation[index] == '{') or (equation[index] == '['):
      character = equation[index]
      position = index
    elif (equation[index] == ')') or (equation[index] == '}') or (equation[index] == ']'):
      if stack.size == 0:
        character = equation[index]
        position = index
      else:
        data = stack.peek()
        character = data[0]
        position = int(data[2])
  else:
    if stack.isEmpty() == False:  # if the stack is not empty even after working with all the brackets
      data = stack.peek()
      character = data[0]
      position = int(data[2])
    else:
      character = None
  
  if (character == '(') or (character == '{') or (character == '['):
    print(f"This expression is NOT correct.\nError at character # {int(position) + 1}. ‘{character}‘- not closed.")
  elif (character == ')') or (character == '}') or (character == ']'):
    print(f"This expression is NOT correct.\nError at character # {index + 1}. ‘{equation[index]}‘- not opened.")
  else:
    print("This expression is correct.")
```

```python
# Linked List based

class stack_underflow_exception(Exception):
  '''raises stack underflow exception'''
  def __init__(self):
    super().__init__("Stack is EMPTY")


class Node:
    def __init__(self, data=None, position = None, bottom=None):
        self.data = data
        self.position = position  # to store the index of the given position
        self.bottom = bottom


class LinkedListStack:
    def __init__(self):
        self.top = None
        self.size = 0

    def push(self, ele, pos):  # will push the position along with the element in node
        new_node = Node(ele, pos, self.top)

        if self.top == None:
            self.top = new_node
        else:
            self.top = new_node

        self.size += 1

    def pop(self):
        if self.size > 0:
            node_to_remove = self.top
            element_removed = node_to_remove.data
            # dont need position in this case

            self.top = node_to_remove.bottom
            self.size -= 1

            # garbage collection
            node_to_remove.data = None
            node_to_remove.position = None
            node_to_remove.bottom = None
            node_to_remove = None

            return element_removed
        else:
            raise stack_underflow_exception

    def peek(self):
        if self.isEmpty():
            raise stack_underflow_exception
        else:
            return f"{self.top.data} {self.top.position}"

    def isEmpty(self):
        if self.size == 0:  # empty
            return True
        else:
            return False


if __name__ == "__main__":
  equation = input("Please enter the equation : ")
  stack = LinkedListStack()
  flag = True
  index = 0
  character = None
  position = None

  while index < len(equation):
    if (equation[index] == '(') or (equation[index] == '{') or (equation[index] == '['):  # for opening brackets
      stack.push(equation[index], index)  # pushing the character along with its index
      index += 1

    elif (equation[index] == ')') or (equation[index] == '}') or (equation[index] == ']'):  # for closing brackets
      try:
        data = stack.peek()
        character = data[0]
        position = data[2]

        if equation[index] == ')':
          if character != '(':
            flag = False
            break
        elif equation[index] == '}':
          if character != '{':
            flag = False
            break
        elif equation[index] == ']':
          if character != '[':
            flag = False
            break
            
        stack.pop()
        index += 1
      except stack_underflow_exception:
        flag = False
        break

    else:  # for characters apart from brackets
        index += 1
  
  # printing part --->
  if flag == False:
    if (equation[index] == '(') or (equation[index] == '{') or (equation[index] == '['):
      character = equation[index]
      position = index
    elif (equation[index] == ')') or (equation[index] == '}') or (equation[index] == ']'):
      if stack.size == 0:
        character = equation[index]
        position = index
      else:
        data = stack.peek()
        character = data[0]
        position = int(data[2])
  else:
    if stack.isEmpty() == False:  # if the stack is not empty even after working with all the brackets
      data = stack.peek()
      character = data[0]
      position = int(data[2])
    else:
      character = None
  
  if (character == '(') or (character == '{') or (character == '['):
    print(f"This expression is NOT correct.\nError at character # {int(position) + 1}. ‘{character}‘- not closed.")
  elif (character == ')') or (character == '}') or (character == ']'):
    print(f"This expression is NOT correct.\nError at character # {index + 1}. ‘{equation[index]}‘- not opened.")
  else:
    print("This expression is correct.")
```
