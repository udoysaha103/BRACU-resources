```python
# task 1
# recursive selection sort

def min_index(array, index):
  if index == len(array)-1:
    return index
  else:
    min_idx = min_index(array, index+1)

    if array[index] < array[min_idx]:
      return index
    else:
      return min_idx


def sel_sort(array, index = 0):
  if index == len(array)-1:
    return
  else:
    min_idx = min_index(array, index)

    array[index], array[min_idx] = array[min_idx], array[index]

    sel_sort(array, index+1)


a = [9,1,5,4,6,3,7,8,9]
print(a)
sel_sort(a)
print(a)
```

```python
# task 2
# recursive insertion sort

def shift(array, index):
  if index == 0:
    return
  else:
    if array[index-1] > array[index]:
      array[index-1], array[index] = array[index], array[index-1]
      shift(array, index-1)
    else:
      return


def inser_sort(array, index = 1):
  if index == len(array):
    return
  else:
    shift(array, index)
    inser_sort(array, index+1)


a = [9,1,5,4,6,3,7,8,9]
print(a)
sel_sort(a)
print(a)
```

```python
# task 3
# bubble sorting on singly linked list

class Node:
  def __init__(self, data = None, next = None):
    self.data = data
    self.next = next

class Linked_List:
  def __init__(self, a):
    self.head = None
    self.len = 0

    if isinstance(a, list):
      current_node = None

      for item in a:
        new_node = Node(item)

        if self.head == None:
          self.head = current_node = new_node
        else:
          current_node.next = new_node
          current_node = new_node
        
        self.len += 1
    else:
      self.head = a
  
  def print(self):
    print(self.head.data, end = "")

    current_node = self.head.next
    while current_node != None:
      print(f" -> {current_node.data}", end = "")
      current_node = current_node.next
    print()


def bubble_sort(linked_list, turn = 1):
  if turn == linked_list.len:
    return
  else:
    current_node = linked_list.head

    while current_node.next != None:
      if current_node.data > current_node.next.data:
        temp = current_node.data
        current_node.data = current_node.next.data
        current_node.next.data = temp

      current_node = current_node.next
    
    bubble_sort(linked_list, turn+1)


a = [9,1,5,4,6,3,7,8,9]
list1 = Linked_List(a)
list1.print()
bubble_sort(list1)
list1.print()
```

```python
# task 5
# insertion sorting on doubly linked list

class Node:
  def __init__(self, data = None, prev = None, next = None):
    self.data = data
    self.prev = prev
    self.next = next

class Linked_List_Doubly:
  def __init__(self, a):
    self.head = None
    self.len = 0

    if isinstance(a, list):
      current_node = None

      for item in a:
        new_node = Node(item, current_node, None)

        if self.head == None:
          self.head = current_node = new_node
        else:
          current_node.next = new_node
          current_node = new_node
        
        self.len += 1
    else:
      self.head = a
  
  def print(self):
    print(self.head.data, end = "")

    current_node = self.head.next
    while current_node != None:
      print(f" <-> {current_node.data}", end = "")
      current_node = current_node.next
    print()


def shift(head):
  if head.prev == None:
    return
  else:
    if head.prev.data > head.data:
      temp = head.data
      head.data = head.prev.data
      head.prev.data = temp

      shift(head.prev)
    else:
      return


def ins_sort(head):
  if head == None:
    return
  else:
    shift(head)
    ins_sort(head.next)


a = [9,1,5,4,6,3,7,8,9]
list1 = Linked_List_Doubly(a)
list1.print()
ins_sort(list1.head)
list1.print()
```

```python
# task 6
# recursive binary search

def bin_search(array, val, l, r):
  if l > r:
    return -1

  m = (l + r) // 2

  if array[m] == val:
    return m
  elif val < array[m]:
    return bin_search(array, val, l, m-1)
  else:
    return bin_search(array, val, m+1, r)


a = [1, 3, 4, 5, 6, 7, 8, 9, 9]
search_val = 3
print(bin_search(a, search_val, 0, len(a)-1))
```
