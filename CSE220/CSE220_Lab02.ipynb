```python
class CircularArray:
  def __init__(self, lin, st, sz):
    # Initializing Variables
    self.start = st
    self.size = sz
    self.cir = [None] * len(lin)
    
    cir_index = self.start

    for lin_index in range(self.size):  # preparing the circular array
      self.cir[cir_index] = lin[lin_index]
      cir_index = (cir_index + 1) % len(self.cir)
    
  
  # Print from index 0 to len(cir) - 1
  def printFullLinear(self): #Easy
    print(self.cir[0], end = "")  # printing the first element of the circular array in its linear form, so, index = 0

    for index in range(1, len(self.cir)):  # printing the rest of the elements of the circular array in its linear form in proper format
      print(f", {self.cir[index]}", end = "")
    
    print()  # newline
  
  
  # Print from start index and total size elements
  def printForward(self): #Easy
    index = self.start

    print(self.cir[index], end = "")  # printing the first element of the circular array, so, index = start
    index = (index + 1) % len(self.cir)

    for offset in range(1, self.size):  # printing the rest of the elements of the circular array in proper format
      print(f", {self.cir[index]}", end = "")
      index = (index + 1) % len(self.cir)
    
    print()  # newline

  
  def printBackward(self): #Easy
    rear = (self.start + self.size - 1) % len(self.cir)
    index = rear

    print(self.cir[index], end = "")  # printing the last element of the circular array, so, index = rear
    index -= 1
    if index < 0:
      index = len(self.cir)-1
    
    for offset in range(1, self.size):  # printing the rest of the elements of the circular array in proper format in reverse order
      print(f", {self.cir[index]}", end = "")
      index -= 1
      if index < 0:
        index = len(self.cir)-1
    
    print()  # newline

  
  # With no null cells
  def linearize(self): #Medium
    new_array = [None] * self.size  # creating a new array with length = size, so that Null values can be avoided
    index = self.start

    for new_index in range(self.size):  # copying the elements of the circular array in forward order
      new_array[new_index] = self.cir[index]

      index = (index + 1) % len(self.cir)
    
    self.cir = new_array  # making the new linear array the actual circular array
    self.start = 0  # linear array has a start value of 0
    
  
  # Do not change the Start index
  def resizeStartUnchanged(self, newcapacity): #Medium
    new_array = [None] * newcapacity  # creating a new array with changed capacity
    new_index = self.start
    old_index = self.start

    for offset in range(self.size):  # copying the elements of the present circular array to the resized array, keeping the start index same
      new_array[new_index] = self.cir[old_index]

      new_index = (new_index + 1) % newcapacity
      old_index = (old_index + 1) % len(self.cir)
    
    self.cir = new_array  # making the new circular array the actual circular array


  # This method will check whether the array is palindrome or not
  def palindromeCheck(self): #Hard
    offset_left = 0
    offset_right = self.size - 1

    flag = True  # to keep record whether the circular array is palindrome or not

    while offset_left < offset_right:  # taking the offset values to compare the mirror elements
      # converting the offset values to linear index
      index1 = (self.start + offset_left) % len(self.cir)
      index2 = (self.start + offset_right) % len(self.cir)

      # comparing if both the indexes are same
      if self.cir[index1] != self.cir[index2]:
        flag = False  # if not same then the circular array is not a palindrome
        break  # not a palindrome, so terminating the comparison
      
      # changing the mirror values accordingly
      offset_left += 1
      offset_right -= 1
    
    # finally printing the result
    if flag == True:
      print("This array is a palindrome")
    else:
      print("This array is NOT a palindrome")


  # This method will sort the values by keeping the start unchanged
  def sort(self):
    for offset_1 in range(self.size - 1):  # selection sort
      index_1 = (self.start + offset_1) % len(self.cir)
      min_index = index_1

      for offset_2 in range(offset_1 + 1, self.size):
        index_2 = (self.start + offset_2) % len(self.cir)

        if self.cir[index_2] < self.cir[min_index]:
          min_index = index_2
      
      if min_index != index_1:
        temp = self.cir[index_1]
        self.cir[index_1] = self.cir[min_index]
        self.cir[min_index] = temp


  # This method will check the given array across the base array and if they are equivalent interms of values return true, or else return false
  def equivalent(self, cir_arr):
    # firstly, if both the arrays are not of same size, then those can not be equivalent
    if self.size != cir_arr.size:
      return False
    
    index_self = self.start
    index_other = cir_arr.start

    for element_no in range(self.size):  # checking the values of corresponding indexes of the two arrays
      if self.cir[index_self] != cir_arr.cir[index_other]:  # if not same, the arrays are not equivalent
        return False
      
      index_self = (index_self + 1) % len(self.cir)
      index_other = (index_other + 1) % len(cir_arr.cir)
    
    return True


  # the method take another circular array and returns a linear array containing the common elements between the two circular arrays.
  def intersection(self, c2):
    length = 0

    for offset1 in range(self.size):  # checking for the number of elements in common
      index1 = (self.start + offset1) % len(self.cir)

      for offset2 in range(c2.size):
        index2 = (c2.start + offset2) % len(c2.cir)

        if self.cir[index1] == c2.cir[index2]:
          length += 1
          break
    
    lin_array = [None] * length  # creating a new array with length = common elements
    lin_index = 0

    for offset1 in range(self.size):  # searching for the common elements and inserting them in the new array
      index1 = (self.start + offset1) % len(self.cir)

      for offset2 in range(c2.size):
        index2 = (c2.start + offset2) % len(c2.cir)

        if self.cir[index1] == c2.cir[index2]:
          lin_array[lin_index] = self.cir[index1]
          lin_index += 1
          break
    
    # finally returning the new array with the elements in common
    return lin_array
```
