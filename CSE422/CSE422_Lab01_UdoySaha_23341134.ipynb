```python
import heapq
import math


def getter():
    f1 = open("Input_file.txt", "r")  # ! Please change the directory according to your file location !

    adj_list = {}
    heuristic_value = {}

    for line in f1.readlines():
        line = line.split()
        heuristic_value[line[0]] = int(line[1])

        if line[0] not in adj_list:
            adj_list[line[0]] = []

        for index in range(2, len(line), 2):
            adj_list[line[0]].append((line[index], int(line[index+1])))


    f1.close()
    return adj_list, heuristic_value


def A_star(adj_list, heuristic_value, start_node, end_node):
    visited = {}
    '''
    Visiting system:
      Unvisited = -1
      Visited but Not Explored = 0
      Explored = 1
    '''
    for node in adj_list.keys():
        visited[node] = -1

    heap = []  # Frontier => Minheap
    heapq.heappush(heap, (0+heuristic_value[start_node], start_node, [start_node], 0))  # f(n), node, path, g(n)
    visited[start_node] = 0

    while len(heap) > 0:
        u = heapq.heappop(heap)

        for v, w in adj_list[u[1]]:
            if visited[v] != 1:  # visiting if not already explored
                heapq.heappush(heap, ((u[3]+w)+heuristic_value[v], v, u[2]+[v], (u[3]+w)))
                visited[v] = 0

        visited[u[1]] = 1

        if u[1] == end_node:
            return u[2], u[3]

    return [], -1  # if there is no path available



if __name__ == "__main__":
    adj_list, heuristic_value = getter()  # scrapping the data
    start_node = input("Start node: ")
    end_node = input("Destination: ")

    path, path_cost = A_star(adj_list, heuristic_value, start_node, end_node)  # implementing A* Search

    if path_cost == -1:
        print("NO PATH FOUND")
    else:
        print('Path:', path[0], end = "")
        for v in path[1:]:
            print(" ->", v, end = "")
        print()
        print("Total distance:", path_cost, "km")
```
