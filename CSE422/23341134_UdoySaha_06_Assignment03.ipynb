```python
# task 1

import math
import random


def maximizer(tree, level, index, alpha, beta):
    if level == 3:  # leaf node
        return tree[index]

    v_prime = None
    v = None

    # for left child
    v_prime = minimizer(tree, level + 1, index * 2, alpha, beta)

    if v == None or v_prime > v:
        v = v_prime
    if v_prime >= beta:
        return v
    if v_prime > alpha:
        alpha = v_prime

    # for right child
    v_prime = minimizer(tree, level + 1, (index * 2) + 1, alpha, beta)

    if v == None or v_prime > v:
        v = v_prime
    if v_prime >= beta:
        return v
    if v_prime > alpha:
        alpha = v_prime

    return v


def minimizer(tree, level, index, alpha, beta):
    if level == 3:  # leaf node
        return tree[index]

    v_prime = None
    v = None

    # for left child
    v_prime = maximizer(tree, level + 1, index * 2, alpha, beta)

    if v == None or v_prime < v:
        v = v_prime
    if v_prime <= alpha:
        return v
    if v_prime < beta:
        beta = v_prime

    # for right child
    v_prime = maximizer(tree, level + 1, (index * 2) + 1, alpha, beta)

    if v == None or v_prime < v:
        v = v_prime
    if v_prime <= alpha:
        return v
    if v_prime < beta:
        beta = v_prime

    return v


def alpha_beta_pruning(tree):
    return maximizer(tree, 0, 1, -math.inf, math.inf)


f1 = open("Input.txt", 'r')  # taking file input
print("Enter your student ID: ", end="")
student_id = f1.readline()
student_id = student_id.replace('0', '8')
print(student_id)
f1.close()

minimum_point = int(student_id[5 - 1])
points_to_win = int(student_id[-1:-3:-1])
maximum_point = math.ceil(points_to_win * 1.5)
shuffle_count = int(student_id[4 - 1])

# implementing array representation of tree data structure
base_tree = [None, 'Max1', 'Min1', 'Min2', 'Max3', 'Max4', 'Max5', 'Max6']  # this is to represent the levels 0, 1 and 2

terminal_nodes = []  # this is for level 3
for index in range(8):
    terminal_nodes.append(random.randint(minimum_point, maximum_point))

game_tree = base_tree + terminal_nodes  # this is the whole tree

optimus_point = alpha_beta_pruning(game_tree)

print("Generated 8 random points between the minimum and maximum point limits:", terminal_nodes)
print("Total points to win:", points_to_win)
print("Achieved point by applying alpha-beta pruning =", optimus_point)

if optimus_point >= points_to_win:
    print("The winner is Optimus Prime")
else:
    print('The Winner is Megatron')





# task 2

def shuffle(leafs):
    for shuffle_times in range(random.randint(1, 4)):  # shuffling for maximum 4 times
        index1, index2 = random.randint(0, 7), random.randint(0, 7)  # picking random indexes
        leafs[index1], leafs[index2] = leafs[index2], leafs[index1]  # shifting the values


shuffled_results = []  # to store the outcome after each shuffle

for each_shuffle in range(shuffle_count):
    shuffle(terminal_nodes)
    shuffled_results.append(alpha_beta_pruning(base_tree + terminal_nodes))  # base_tree + terminal_nodes = actual game tree

times_won = 0
for each_result in shuffled_results:
    if each_result >= points_to_win:
        times_won += 1

print("\nAfter the shuffle:")
print("List of all points values from each shuffle:", shuffled_results)
print("The maximum value of all shuffles:", max(shuffled_results))
print(f"Won {times_won} times out of {shuffle_count} number of shuffles")
```
