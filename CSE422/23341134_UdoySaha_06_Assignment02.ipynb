```python
import random

def getter():  # Scrapping the input
    f = open('Input.txt', 'r')

    n, target = list(map(int, f.readline().split()))

    players = []
    averages = []

    for line in f.readlines():
        line = line.strip().split(' ')
        if len(line) == 2:
            players.append(line[0])
            averages.append(int(line[1]))
        elif len(line) == 1:
            index = 0
            while index < len(line[0]):
                if line[0][index].isalpha():
                    index += 1
                else:
                    break
            players.append(line[0][0: index])
            averages.append(int(line[0][index: ]))
    f.close()
    return n, target, players, averages


def parent_generator():  # Generating the initial population
    parents = []

    while len(parents) < 15:
        new_parent = []

        for index in range(n):  # Randomly generating a parent
            new_parent.append(random.randint(0, 1))

        if (sum(new_parent) > 0) and (new_parent not in parents):
            parents.append(new_parent)

    return parents


def fitness(member):  # Function to calculate the fitness
    summation = 0
    for index in range(len(member)):
        if member[index]:
            summation += averages[index]

    return abs(target - summation)  # The less the number is, the fitter the chromosome is. Perfectly fit = 0.


def crossover(parent1, parent2):  # Function to perform crossover between two parents
    crossing_point = n // 2

    child1 = parent1[0: crossing_point] + parent2[crossing_point: ]
    child2 = parent2[0: crossing_point] + parent1[crossing_point: ]

    return child1, child2


def mutate(member):  # Mutating one random gene of the chromosome
    index = random.randint(0, n-1)
    if member[index] == 0:
        member[index] = 1
    else:
        member[index] = 0


def genetic_algo():  # Genetic algorithm simulator function
    parents = parent_generator()  # Generated the initial chromosomes
    children = []

    for iteration_count in range(500):  # !! Changeable maximum iteration_count count !!
        while len(parents) > 1:
            index1 = random.randint(0, len(parents)-1)
            index2 = random.randint(0, len(parents)-1)

            while index1 == index2:
                index2 = random.randint(0, len(parents) - 1)

            # randomly selected parents from the parent population
            parent1 = parents[index1]
            parent2 = parents[index2]

            # Performing Crossover
            child1, child2 = crossover(parent1, parent2)

            # Performing Mutation
            mutation_threshold = 0.5  # !! Changeable mutation-threshold !!

            if random.random() >= mutation_threshold:
                mutate(child1)
            if random.random() >= mutation_threshold:
                mutate(child2)

            if fitness(child1) == 0:  # Checking whether it is the perfectly fit chromosome
                # print(iteration_count)  # !! Uncomment if needed !! (Iteration count to reach perfectly fit chromosome)
                return child1
            if fitness(child2) == 0:  # Checking whether it is the perfectly fit chromosome
                # print(iteration_count)  # !! Uncomment if needed !! (Iteration count to reach perfectly fit chromosome)
                return child2

            # Discarding the child(ren) if there is no gene in the chromosome
            if sum(child1) == 0:
                child1 = None
            if sum(child2) == 0:
                child2 = None

            # Including the child(ren) in the child population
            if child1:
                children.append(child1)
            if child2:
                children.append(child2)

            # Excluding the mated parents from the parent population
            parents.remove(parent1)
            parents.remove(parent2)

        if len(parents) == 1:  # if only 1 parent remained
            if fitness(parents[0]) == 0:  # Checking whether it is the perfectly fit chromosome
                # print(iteration_count)  # !! Uncomment if needed !! (Iteration count to reach perfectly fit chromosome)
                return parents[0]

        parents = children  # current child population becomes the new parent population
        children = []  # new empty parent population

    return -1  # No perfectly fit chromosome found


if __name__ == '__main__':
    n, target, players, averages = getter()
    player_combination = genetic_algo()

    print(players)
    if player_combination == -1:
        print(-1)
    else:
        print("".join(map(str, player_combination)))
```
