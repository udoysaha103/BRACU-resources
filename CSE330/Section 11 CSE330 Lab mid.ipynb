Q3. We know standard format of polynomial of degree 4 :

P(x)= a0 + a1*x + a2*x^2 + a3*x^3 + a4*x^4

But you have to create a different version of polynomial which is given below:

P(x)= a4 + a3*x + a2*x^2 + a1*x^3 + a0*x^4

Then find approximate value of your created polynomial at x=4

Coefficients are given [a0,a1,a2,a3,a4]=[1,2,3,4,5]

```python
#Q4
coeff = [1, 2, 3, 4, 5]
n = len(coeff)-1
px = 0
x = 4
#use a loop to find value of px at x
for i in range(n+1):
  px += coeff[i] * (x ** (n - i))

print(px)
```

Q.5 The formula for Lagrange basis is as follows. Implement the the formula.

\begin{equation}
l_k(x) = \prod_{j=0, j\neq k}^{n} \frac{x-x_j}{x_k-x_j}. \tag{2}
\end{equation}

return the value of lagrange basis in the given function.


```python
def l(k, x, x1):
  prod = 1

  for j in range(len(x)):
    if j == k:
      continue
    
    prod *= (x1 - x[j]) / (x[k] - x[j])
  
  return prod
```

```python
x = [3, 4, 5, 6, 7]
x1 = 9
print(l(2, x, x1))
```
