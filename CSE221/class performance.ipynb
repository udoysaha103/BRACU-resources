```python
# connected component finder in a graph
# Algorithm: Calling DFS for every vertex and checking how many times DFS is being called. The count of DFS call is the number of connected components in the graph.

def DFS(adj_list, node, visit_list):  # regular recursive DFS algorithm
  visit_list[node] = 0  # the node is just visited, so visit = 0

  for i in adj_list[node]:  # checking dfs call for every adjacent vertex
    if visit_list[i] == -1:  # calling dfs only when unvisited
      DFS(adj_list, i, visit_list)
  
  visit_list[node] = 1  # after the node is explored, visit = 1


n = 5  # input of n (Hardcoded)
edges = [[0,1],[1,2],[3,4]]  # input of all the edges (Hardcoded)

adj_list = {}
for i in range(n):
  adj_list[i] = []

for u, v in edges:  # creating the adjacency list
  adj_list[u].append(v)
  adj_list[v].append(u)

visited = [-1] * n
# -1 = unvisited
# 0 = visited but not explored
# 1 = visited and explored

count = 0  # count of connected components
for i in range(n):  # checking if DFS is callable for all the vertices
  if visited[i] == -1:  # calling DFS only when unvisited
    count += 1
    DFS(adj_list, i, visited)

print(count)
```
