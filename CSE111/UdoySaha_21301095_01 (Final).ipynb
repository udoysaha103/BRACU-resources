```python
#1

my_dict = {}
inputno = 0

while True:
    user_input = input("Please enter: ")
    if user_input.lower() == "stop":
        break

    inputno += 1
    brand, price = user_input.split()

    watch_id = str(inputno) + '_' + brand[:2] + '_' + price

    if brand not in my_dict.keys():
        my_dict[brand] = [watch_id]
    else:
        my_dict[brand].append(watch_id)

print(my_dict)
```

```python
#2

class Laptop:
    total_laptop = 0

    def __init__(self, model, price = "Not Set"):
        self.model = model
        self.price = price

        print("Laptop created.")
        print(f"Model: {self.model}")
        print(f"Price: {self.price}")

        Laptop.total_laptop += 1

    def setPrice(self, new_price):
        self.price = new_price

class Store:
    def __init__(self, name = "Ryans"):
        self.name = name
        self.total = 0
        self.laptops = {}

    def add_laptop(self, *all_laptops):
        for laptop in all_laptops:
            if laptop.price != "Not Set":
                self.laptops[laptop.model] = laptop.price
                self.total += 1
                print(f"{laptop.model} model has been added to {self.name} store.")
            else:
                print(f"You can not add {laptop.model} without price")

    def sell_laptop(self, model):
        if model not in self.laptops.keys():
            print(f"{model} is not available at the {self.name} store")
        else:
            print(f"{model} has been sold at the price {self.laptops[model]}")
            self.total -= 1
            self.laptops.pop(model)

    def __str__(self):
        s = f"Store Name: {self.name}\nList of available Laptops:"
        for model, price in self.laptops.items():
            s += f"\nModel Name: {model} --> Price: {price}"
        s += f"\nTotal Laptop in Stock: {len(self.laptops.keys())}"
        s += f"\nTotal Price of Stocked Laptops: {sum(list(self.laptops.values()))}"

        return s


print("Total Laptop Created:",Laptop.total_laptop)
print("1.=========================================")
s1 = Store()
p1 = Laptop("Lenovo B4180", 55000)
print("2.=========================================")
p2 = Laptop("HP Pavilion")
print("3.=========================================")
s1.add_laptop(p1, p2)
print("4.=========================================")
p2.setPrice(77000)
s1.add_laptop(p2)
print("5.=========================================")
s1.sell_laptop("HP Probook")
print("6.=========================================")
print(s1)
print("7.=========================================")
s2 = Store("Star Tech")
p3 = Laptop("HP Probook", 58000)
print("8.=========================================")
p4 = Laptop("Lenovo Thinkpad", 60000)
print("9.=========================================")
p5 = Laptop("HP Elite", 80000)
print("10.=========================================")
s2.add_laptop(p3, p4, p5)
print("11.=========================================")
print(s2)
print("12.=========================================")
s2.sell_laptop("HP Probook")
print("13.=========================================")
print(s2)
print("14.=========================================")
print("Total Laptop Created:",Laptop.total_laptop)
```

```python
#3

class Person:
    def __init__(self, name, contact):
        self.name = name
        self.contact = contact

    def __str__(self):
        s = f"Name: {self.name}\nContact: {self.contact}"
        return s

# ----------------------------------------------------------------------------------
# Write your code here

class Student(Person):
    def __init__(self, name, contact, student_id, department):
        super().__init__(name, contact)
        self.student_id = student_id
        self.department = department
        self.courses = {}

    def add_courses(self, *all_courses):
        for course in all_courses:
            dpt = course[:3]

            if dpt not in self.courses.keys():
                self.courses[dpt] = [course]
            else:
                self.courses[dpt].append(course)

    def __str__(self):
        s = "Student Details:\n"
        s += super().__str__()
        s += f"\nID: {self.student_id}\n\
Department: {self.department}\n\
Courses List:\n"
        s += str(self.courses)

        return s

    def show_course_list(self, dpt):
        if dpt in self.courses.keys():
            print(f"Courses from {dpt} department:")

            for course in self.courses[dpt]:
                print(course)
        else:
            print(f"{self.name} did not take any course from {dpt} department")


class Lawyer(Person):
    def __init__(self, name, contact, area):
        super().__init__(name, contact)
        self.area = area
        self.clients = {}

    def add_client(self, *all_clients):
        for client in all_clients:
            client_type = client.split()[-1]

            if client_type not in self.clients.keys():
                self.clients[client_type] = [client]
            else:
                self.clients[client_type].append(client)

    def __str__(self):
        s = "Lawyer Details:\n"
        s += super().__str__()
        s += f"\nArea: {self.area}\n\
Client List: "
        s += str(self.clients)

        return s

# ----------------------------------------------------------------------------------


john_cena = Student("John Cena", "9752325", 100005, "CSE")
john_cena.add_courses("CSE111 Programming Language II", "CSE230 Discrete Mathematics", "EEE101 Electrical Circuits 1")
print("1.=========================================================")
print(john_cena)
print("2.=========================================================")
john_cena.show_course_list("CSE")
print("3.=========================================================")
john_cena.show_course_list("ECE")
print("4.=========================================================")
john_cena.add_courses("ARC101 Design I: Basic Design")
print("5.=========================================================")
print(john_cena)
print("6.=========================================================")
saul_goodman = Lawyer("Saul Goodman", "5885858", "Dhaka")
saul_goodman.add_client("Jesse Pinkman Criminal", "Walter White Criminal", "John Cena Civil", "Naruto Uzumaki Civil")
print("7.=========================================================")
print(saul_goodman)
```
