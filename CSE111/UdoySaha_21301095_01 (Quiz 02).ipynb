```python
class TechShop:
    def __init__(self, shop_name):
        self.shop_name = shop_name
        self.products = []

    def addProduct(self, *products):
        for product in products:
            if product.product_code != None:
                self.products.append(product)
                print(f"Product with Code: {product.product_code} added")
            else:
                print("A product without code cannot be added.")

    def printProducts(self):
        for product in self.products:
            print(f"Product name:{product.product_name} Category:{product.product_category} Product Code:{product.product_code}")


class Product:
    def __init__(self, product_name, product_category, product_code = None):
        self.product_name = product_name
        self.product_category = product_category
        self.product_code = product_code

    def setCode(self, new_code):
        self.product_code = new_code


if __name__ == "__main__":
    obj = TechShop("Star Tech")
    p1 = Product("Core i7 11700", "Processor", 16039)
    print("=================================")
    obj.addProduct(p1)
    print("=================================")
    obj.printProducts()
    print("=================================")
    p2 = Product("Ryzen 5900X", "Processor")
    p3 = Product("ASRock B550 Taichi", "Motherboard", 18080)
    print("=================================")
    obj.addProduct(p2, p3)
    print("=================================")
    obj.printProducts()
    print("=================================")
    p2.setCode(17699)
    print("=================================")
    obj.addProduct(p2)
    print("=================================")
    obj.printProducts()
    print("=================================")
    p4 = Product("ASUS ROG Strix Z390-F", "Motherboard", 8694)
    p5 = Product("ASUS ROG Strix RTX3080", "GPU", 17071)
    print("=================================")
    obj.addProduct(p4, p5)
    print("=================================")
    obj.printProducts()
    print("=================================")
```
