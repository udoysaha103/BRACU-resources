```python
#1

class Calculator:
    def __init__(self):
        print("Let's Calculate!")

    def add(self, first_val, second_val):
        return first_val + second_val

    def subtract(self, first_val, second_val):
        return first_val - second_val

    def multiply(self, first_val, second_val):
        return first_val * second_val

    def divide(self, first_val, second_val):
        return first_val / second_val


if __name__ == "__main__":
    my_calculator = Calculator()

    first_val = int(input("Value 1: "))
    operator = input("Operator: ")
    second_val = int(input("Value 2: "))

    result = None

    if operator == '+':
        result = my_calculator.add(first_val, second_val)
    elif operator == '-':
        result = my_calculator.subtract(first_val, second_val)
    elif operator == '*':
        result = my_calculator.multiply(first_val, second_val)
    elif operator == '/':
        result = my_calculator.divide(first_val, second_val)

    print("Result:", result)
```

```python
#2

class Customer:
    def __init__(self, name):
        self.name = name

    def greet(self, name = None):
        if name != None:
            print(f"Hello {self.name}!")
        else:
            print("Hello!")

    def purchase(self, *args):
        print(f"{self.name}, you purchased {len(args)} item(s):")

        for item in args:
            print(item)


if __name__ == "__main__":
    customer_1 = Customer("Sam")
    customer_1.greet()
    customer_1.purchase("chips", "chocolate", "orange juice")
    print("-----------------------------")
    customer_2 = Customer("David")
    customer_2.greet("David")
    customer_2.purchase("orange juice")
```

```python
#3

class Panda:
    def __init__(self, name, gender, age):
        self.name = name
        self.gender = gender
        self.age = age

    def sleep(self, hours = None):
        if hours == None:
            return f"{self.name}'s duration is unknown thus should have only bamboo leaves"
        else:
            food = None

            if (hours >= 3) and (hours <= 5):
                food = "Mixed Veggies"
            elif (hours >= 6) and (hours <= 8):
                food = "Eggplant & Tofu"
            elif (hours >= 9) and (hours <= 11):
                food = "Broccoli Chicken"

            return f"{self.name} sleeps {hours} hours daily and should have {food}"


if __name__ == "__main__":
    panda1 = Panda("Kunfu", "Male", 5)
    panda2 = Panda("Pan Pan", "Female", 3)
    panda3 = Panda("Ming Ming", "Female", 8)
    print("{} is a {} Panda Bear who is {} years old ".format(panda1.name,panda1.gender,panda1.age))
    print("{} is a {} Panda Bear who is {} years old ".format(panda2.name,panda2.gender,panda2.age))
    print("{} is a {} Panda Bear who is {} years old ".format(panda3.name,panda3.gender,panda3.age))
    print("===========================")
    print(panda2.sleep(10))
    print(panda1.sleep(4))
    print(panda3.sleep())
```

```python
#13

class Batsman:
    def __init__(self, *player_info):
        if len(player_info) == 3:
            self.name = player_info[0]
            self.score = player_info[1]
            self.ball = player_info[2]
        elif len(player_info) == 2:
            self.name = "New Batsman"
            self.score = player_info[0]
            self.ball = player_info[1]

    def printCareerStatistics(self):
        print(f"Name: {self.name}")
        print(f"Runs Scored: {self.score} , Balls Faced: {self.ball}")

    def battingStrikeRate(self):
        return (self.score / self.ball) * 100

    def setName(self, new_name):
        self.name = new_name


if __name__ == "__main__":
    b1 = Batsman(6101, 7380)
    b1.printCareerStatistics()
    print("============================")
    b2 = Batsman("Liton Das", 678, 773)
    b2.printCareerStatistics()
    print("----------------------------")
    print(b2.battingStrikeRate())
    print("============================")
    b1.setName("Shakib Al Hasan")
    b1.printCareerStatistics()
    print("----------------------------")
    print(b1.battingStrikeRate())
```

```python
#20

class Author:
    def __init__(self, name = None):
        self.name = name
        self.books = {}

    def addBook(self, *book_info):
        if self.name == None:
            print("A book can not be added without author name")
        else:
            if book_info[1] in self.books.keys():
                self.books[book_info[1]].append(book_info[0])
            else:
                self.books.update({book_info[1]: [book_info[0]]})

    def setName(self, new_name):
        self.name = new_name

    def printDetail(self):
        number_of_books = 0
        for category in self.books.keys():
            number_of_books += len(self.books[category])

        print("Number of Book(s):", number_of_books)

        print(f"Author Name: {self.name}")

        for category, books in self.books.items():
            print(f"{category}: {books[0]}", end = "")

            for book in books[1:]:
                print(",", book, end = "")

            print()


if __name__ == "__main__":
    a1 = Author()
    print("=================================")
    a1.addBook("Ice", "Science Fiction")
    print("=================================")
    a1.setName("Anna Kavan")
    a1.addBook("Ice", "Science Fiction")
    a1.printDetail()
    print("=================================")
    a2 = Author("Humayun Ahmed")
    a2.addBook("Onnobhubon", "Science Fiction")
    a2.addBook("Megher Upor Bari", "Horror")
    print("=================================")
    a2.printDetail()
    a2.addBook("Ireena", "Science Fiction")
    print("=================================")
    a2.printDetail()
    print("=================================")
```

```python
#21

class Hospital:
    doctors = {}
    patients = {}

    def __init__(self, *info):
        if len(info) == 3:
            self.name = info[1]
            self.speciality = info[2]
            self.doctor_id = None
        else:
            self.name = info[1]
            self.age = info[2]
            self.phone = info[3]
            self.patient_id = None

    def addDoctor(self, doctor_id):
        self.doctor_id = doctor_id

        self.doctors[self.doctor_id] = [self.name, self.speciality]

    def addPatient(self, patient_id):
        self.patient_id = patient_id

        self.patients[self.patient_id] = [self.name, self.age, self.phone]

    def getDoctorByID(self, doctor_id):
        return f"Doctor's ID: {doctor_id}\nName: {self.doctors[doctor_id][0]}\nSpeciality: {self.doctors[doctor_id][1]}"

    def getPatientByID(self, patient_id):
        return f"Patient's ID: {patient_id}\nName: {self.patients[patient_id][0]}\nAge: {self.patients[patient_id][1]}\nPhone no.: {self.patients[patient_id][2]}"

    def allDoctors(self):
        print("All Doctors:")
        print("Number of Doctors:", len(self.doctors.keys()))
        print(self.doctors)

    def allPatients(self):
        print("All Patients:")
        print("Number of Patients:", len(self.patients.keys()))
        print(self.patients)


if __name__ == "__main__":
    d1 = Hospital("Doctor", "Samar Kumar", "Neurologist")
    d1.addDoctor("1d")
    print(d1.getDoctorByID("1d"))
    print("=================================")
    p1 = Hospital("Patient", "Kashem Ahmed", 35, 12345)
    p1.addPatient("1p")
    print(d1.getPatientByID("1p"))
    print("=================================")
    p2 = Hospital("Patient", "Tanina Haque", 26, 33456)
    p2.addPatient("2p")
    print(d1.getPatientByID("2p"))
    print("=================================")
    d1.allDoctors()
    p1.allPatients()
```

```python
#22

class Vaccine:
    def __init__(self, name, origin, frequency):
        self.name = name
        self.origin = origin
        self.frequency = frequency

    def compare(self, other):
        if self.name == other.name:
            return True
        else:
            return False


class Person:
    def __init__(self, name, age, type = "General Citizen"):
        self.name = name
        self.age = age
        self.type = type
        self.vaccine = None
        self.first_dose = None
        self.second_dose = None

    def pushVaccine(self, pushing_vaccine, dose = "1st Dose"):
        if self.type == "Student" or self.age >= 25:
            if self.vaccine == None:
                self.vaccine = pushing_vaccine
                self.first_dose = "Given"
                self.second_dose = f"Please come after {self.vaccine.frequency} days"

                print(f"1st dose done for {self.name}")
            else:
                if self.vaccine.compare(pushing_vaccine):
                    self.second_dose = "Given"

                    print(f"2nd dose done for {self.name}")
                else:
                    print(f"Sorry {self.name}, you canâ€™t take 2 different vaccines")
        else:
            print(f"Sorry {self.name}, Minimum age for taking vaccines is 25 years now.")

    def showDetail(self):
        print(f"Name: {self.name} Age: {self.age} Type: {self.type}")
        print(f"Vaccine name: {self.vaccine.name}")
        print(f"1st dose: {self.first_dose}")
        print(f"2nd dose: {self.second_dose}")


if __name__ == "__main__":
    astra = Vaccine("AstraZeneca", "UK", 60)
    modr = Vaccine("Moderna", "UK", 30)
    sin = Vaccine("Sinopharm", "China", 30)
    p1 = Person("Bob", 21, "Student")
    print("=================================")
    p1.pushVaccine(astra)
    print("=================================")
    p1.showDetail()
    print("=================================")
    p1.pushVaccine(sin, "2nd Dose")
    print("=================================")
    p1.pushVaccine(astra, "2nd Dose")
    print("=================================")
    p1.showDetail()
    print("=================================")
    p2 = Person("Carol", 23, "Actor")
    print("=================================")
    p2.pushVaccine(sin)
    print("=================================")
    p3 = Person("David", 34)
    print("=================================")
    p3.pushVaccine(modr)
    print("=================================")
    p3.showDetail()
    print("=================================")
    p3.pushVaccine(modr, "2nd Dose")
```
