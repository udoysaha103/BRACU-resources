```python
#1

user_input = input("Please enter a string : ")

my_dict = {}

for character in user_input:
    if (character >= 'A') and (character <= 'Z'):
        character = chr(ord(character) + 32)

    if character not in my_dict:
        my_dict[character] = 1
    else:
        my_dict[character] += 1

frequencies = list(my_dict.values())

maximum = frequencies[0]
for number in frequencies[1:]:
    if number > maximum:
        maximum = number

second_max = 0
for number in frequencies:
    if (number > second_max) and (number != maximum):
        second_max = number

index = 0

while index < len(user_input):
    character = user_input[index]

    if (character >= 'A') and (character <= 'Z'):
        character = chr(ord(character) + 32)

    if my_dict[character] == second_max:
        print(chr(ord(character) - 32))
        break

    index += 1
```

```python
#2

class SDS:
    def __init__(self, year):
        self.year = year
        self.no_cse = 0
        self.no_mns  = 0
        self.cse_students = {}
        self.mns_students = {}

        print(f"School of Data and Sciences {self.year}! Only CSE and MNS department students can be added.")

    def print_info(self):
        print(f"Number of CSE students: {self.no_cse}\nNumber of MNS students: {self.no_mns}\nCSE Students {self.cse_students}\nMNS Students {self.mns_students}")

    def addStudent(self, *students):
        for student in students:
            if student.department == "MNS":
                self.no_mns += 1
                self.mns_students[student.student_id] = [student.name, student.cgpa, student.courses]
            elif student.department == "CSE":
                self.no_cse += 1
                self.cse_students[student.student_id] = [student.name, student.cgpa, student.courses]
            else:
                print(f"{student.department} department student cannot be added to SDS")


class Student:
    def __init__(self, name, student_id, department = "MNS", courses = [], cgpa = 0):
        self.name = name
        self.student_id = student_id
        self.department = department
        self.courses = courses
        self.cgpa = cgpa

    def print_info(self):
        print(f"Name: {self.name}\nID: {self.student_id}\nDepartment: {self.department}\nCourses: {self.courses}\nCGPA: {self.cgpa}")


if __name__ == "__main__":
    sds21 = SDS("2021")
    print("*********************************************")
    sds21.print_info()
    print("1.================================")
    s1 = Student("Maliha", 21101001)
    s1.print_info()
    print("2.================================")
    sds21.addStudent(s1)
    sds21.print_info()
    print("3.================================")
    s2 = Student("Subha", 21301001, "CSE", ["ENG091", "MAT092"], 3.48)
    s2.print_info()
    print("4.================================")
    sds21.addStudent(s2)
    sds21.print_info()
    print("5.================================")
    s3 = Student("Samiul", 21305001, "MNS", ["ENG091", "MAT092", "PHY111"], 3.01)
    s4 = Student("Mohsina", 21305023)
    print("6.================================")
    s3.print_info()
    print("7.================================")
    s4.print_info()
    print("8.================================")
    sds21.addStudent(s3, s4)
    print("9.================================")
    sds21.print_info()
    print("10.================================")
    s5 = Student("Nayla", 16101288, "BBS")
    sds21.addStudent(s5)
    print("11.================================")
    sds21.print_info()
```
